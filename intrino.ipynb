```python
import intrinio_sdk
from intrinio_sdk.rest import ApiException
from datetime import datetime

# Configure the API key
API_KEY = "OjMxZDhjNTYyNjcxZmM0NzNiZDQ1ZWIxZWZmZmFmNzYz"
intrinio_sdk.ApiClient().configuration.api_key['api_key'] = API_KEY

# Initialize API client
security_api = intrinio_sdk.SecurityApi()

# Magnificent 7 tickers
MAG_7 = ["GOOGL", "AMZN", "AAPL", "META", "MSFT", "NVDA", "TSLA"]

def get_realtime_price(ticker):
    """Fetch real-time price for a given ticker."""
    try:
        response = security_api.get_security_realtime_price(ticker, source="iex")
        print(f"\nReal-Time Price for {ticker}:")
        print(f"Last Price: {response.last_price}")
        print(f"Last Time: {response.last_time}")
        print(f"Source: {response.source}")
    except ApiException as e:
        print(f"\nError accessing real-time price for {ticker}: {e}")

def get_security_info(ticker):
    """Fetch basic security info for a given ticker."""
    try:
        response = security_api.get_security_by_id(ticker)
        print(f"\nSecurity Info for {ticker}:")
        print("Full Response:", vars(response))
        print(f"Name: {response.name}")
        print(f"Ticker: {response.ticker}")
        if hasattr(response, 'listing_exchange_mic'):
            print(f"Exchange MIC: {response.listing_exchange_mic}")
        else:
            print("No exchange info available.")
    except ApiException as e:
        print(f"\nError accessing security info for {ticker}: {e}")

def get_historical_prices(ticker):
    """Attempt to fetch historical prices for a given ticker."""
    try:
        response = security_api.get_security_stock_prices(ticker, start_date="2024-01-01", end_date="2024-12-31")
        print(f"\nHistorical Prices for {ticker} (first 5 days):")
        for price in response.stock_prices[:5]:
            print(f"Date: {price.date}, Close: {price.close}")
    except ApiException as e:
        print(f"\nError accessing historical prices for {ticker}: {e}")

def explore_mag_7():
    """Explore data for the Magnificent 7 stocks."""
    print(f"Exploring Intrinio API Access for Magnificent 7 as of {datetime.now()}...\n")
    for ticker in MAG_7:
        print(f"===== {ticker} =====")
        get_realtime_price(ticker)
        get_security_info(ticker)
        get_historical_prices(ticker)
    print("\nExploration complete. Check the output for available data.")

if __name__ == "__main__":
    explore_mag_7()
```
